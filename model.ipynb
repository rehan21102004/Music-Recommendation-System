```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
```

```python
df=pd.read_csv("Bollywood_Songs_With_Album_Genre.csv")
```

```python
df['total'] =(
    df['song_name'].astype(str) + " " +
    df['artist'].astype(str) + " " +
    df['album'].astype(str) + " " +
    df['genre'].astype(str)   
)
```

```python
df['total']
```

```python
tfidf = TfidfVectorizer(stop_words='english')
vectors = tfidf.fit_transform(df['total'])
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommendation(song):
    song=song.lower()

    try:
        index=df[df['song_name'].str.lower()== song].index[0]
    except:
        return []
    distance = similarity[index]
    song_list = sorted(list(enumerate(distance)),
                       reverse=True,key=lambda x:x[1])[1:6]
    
    recommend = []
    for i in song_list:
        recommend.append(df.iloc[i[0]]["song_name"])

    return recommend
```

```python
pickle.dump(df,open("songs_df.pkl","wb"))
pickle.dump(similarity,open("similarity.pkl","wb"))
pickle.dump(tfidf,open("tfidf.pkl","wb"))
```
